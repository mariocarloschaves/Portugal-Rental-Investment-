```python
from __future__ import annotations

import json
import runpy
import sys
from pathlib import Path

import pandas as pd
```

```python
cwd = Path.cwd().resolve()

candidates = [
    cwd,
    cwd.parent,
    cwd.parent.parent,
    cwd / "deployment_models",
    cwd / "code files" / "ml" / "deployment_models",
    cwd.parent / "deployment_models",
    cwd.parent / "ml" / "deployment_models",
    cwd.parent / "code files" / "ml" / "deployment_models",
    Path(r"C:\Users\mario\Documents\DSMLAISL LEARNING PATH\Portugal-Rental-Investment-\code files\ml\deployment_models"),
]

DEPLOYMENT_DIR = None

for candidate in candidates:
    if (candidate / "occupancy_deployment_model_v3.py").exists():
        DEPLOYMENT_DIR = candidate
        break

if DEPLOYMENT_DIR is None:
    print("Current working directory:", cwd)
    print("\nChecked folders:")
    for candidate in candidates:
        print(candidate)
    raise FileNotFoundError("Could not find occupancy_deployment_model_v3.py")

ML_DIR = DEPLOYMENT_DIR.parent

for path in [DEPLOYMENT_DIR, ML_DIR]:
    if str(path) not in sys.path:
        sys.path.insert(0, str(path))

OCCUPANCY_SCRIPT = DEPLOYMENT_DIR / "occupancy_deployment_model_v3.py"

print("DEPLOYMENT_DIR:", DEPLOYMENT_DIR)
print("ML_DIR:", ML_DIR)
print("OCCUPANCY_SCRIPT:", OCCUPANCY_SCRIPT)
```

```python
from gold_ml_modeling import DEPLOYMENT_MODEL_DIR

model_path = DEPLOYMENT_MODEL_DIR / "occupancy_deployment_model_v3.joblib"
metadata_path = DEPLOYMENT_MODEL_DIR / "occupancy_deployment_model_v3_metadata.json"

print("Model path:", model_path)
print("Metadata path:", metadata_path)
```

```python
runpy.run_path(str(OCCUPANCY_SCRIPT), run_name="__main__")
```
