```python
from pathlib import Path
import sys
import json
import runpy
import pandas as pd
```

```python
THIS_DIR = Path.cwd().resolve()

candidates = [
    THIS_DIR,
    THIS_DIR.parent,
    THIS_DIR / "code files" / "ml" / "deployment_models",
    THIS_DIR.parent / "deployment_models",
]

DEPLOYMENT_DIR = None
for candidate in candidates:
    if (candidate / "price_deployment_model_v2.py").exists():
        DEPLOYMENT_DIR = candidate
        break

if DEPLOYMENT_DIR is None:
    raise FileNotFoundError("Could not find price_deployment_model_v2.py")

PRICE_SCRIPT = DEPLOYMENT_DIR / "price_deployment_model_v2.py"

print("DEPLOYMENT_DIR:", DEPLOYMENT_DIR)
print("PRICE_SCRIPT:", PRICE_SCRIPT)
```

```python
runpy.run_path(str(PRICE_SCRIPT), run_name="__main__")
```
